```python
X = df.iloc[:,0:2]
scaler = StandardScaler()
print(scaler.fit(X))
print(scaler.mean_)
X=scaler.transform(X)
```

```python
X = pd.DataFrame(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train0 = X_train.iloc[:,0].to_numpy()
```

```python
X_train1 = X_train.iloc[:, 1].to_numpy()
```

```python
X_train1
```

```python
import math
```

```python
b0 = 0
b1=0
b2=0
a = 0.001
y_train = y_train.to_numpy()
```

```python
for i in range(0,10000):
    for j in range(0,80):
        ypred = b0 + b1*X_train0[j] + b2*X_train1[j]
        pred  = 1/(1+np.exp(-ypred))
        b0=b0+a*(y_train[j]-pred)*pred*(1-pred)*1
        b1=b1+a*(y_train[j]-pred)*pred*(1-pred)*X_train0[j]
        b2=b2+a*(y_train[j]-pred)*pred*(1-pred)*X_train1[j]
```

```python
print(b0,b1,b2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ypred1 = b0 + b1*X_test0 + b2*X_test1
pred1  = 1/(1+np.exp(-ypred1))
np.round(pred1)
```

```python
accuracy_score(y_test,np.round(pred1))
```
